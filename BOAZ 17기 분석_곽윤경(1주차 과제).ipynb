### 6. grade별로 평균 가격을 보여주세요.

```python
grade_mean= file.groupby('grade')['price'].agg([('grade_mean','mean')])
display(grade_mean)
```

### 7. 6번에서 만든 데이터를 원본 데이터와 merge해주세요.

```python
pd.merge(file, grade_mean, left_on='grade', right_on='grade', how='inner')
```

### 8. Date Column을 이용해 년, 월, 일에 해당하는 3개의 새로운 Column을 만들어주세요.

```python
file['year'] = file.date.apply(lambda x : x[:4])
file['month'] = file.date.apply(lambda x : x[4:6])
file['day'] = file.date.apply(lambda x : x[6:8])
display(file.head())
```
