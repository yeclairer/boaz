### 건축 년도와 가격의 상관 관계를 보기 위해 
### 재건축 년도와 원래 건축 년도를 비교해서 마지막 건축 년도를 추가적인 변수로 생성한다.

```python
re_built = data['yr_renovated']
re_built = np.array(re_built)

built = data['yr_built']
built = np.array(built)
```

```python
last_built = []
for i in range(len(built)):
    if re_built[i] != 0:
        last_built.append(re_built[i])
    else:
        last_built.append(built[i])
```

```python
col_name = ['last_built']
last_built = pd.DataFrame(last_built, columns=col_name)

data['last_built'] = last_built
```

### 새로 생성한 변수인 마지막 건축 년도와 price 사이에는 유의미한 상관관계를 찾기 어려움을 확인할 수 있다.

##### 정성경

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.rc('font', family='malgun gothic')
plt.rc('axes', unicode_minus=False)
```

```python
data['rooms'] = (data.bedrooms * data.bathrooms/10).astype(int)
groups = data.groupby('rooms')

fig, ax = plt.subplots()

for name, group in groups:
    ax.plot(group.grade, 
            group.price, 
            marker='o', 
            linestyle='',
            markersize=5,
            label=name)

ax.legend(loc='upper left') # legend position
plt.title('Scatter plot of rooms(bedrooms*bathrooms/10)', fontsize=14)
plt.xlabel('Grade', fontsize=14)
plt.ylabel('Price', fontsize=14)
plt.show()
```

```python
# 우선적으로 Grade와 Price가 비례하게 나타나는 것을 확인할 수 있다.
# 추가적으로 Bedrooms * Bathrooms / 10 를 int로 반환하여 scatter plot한 결과...
# 낮은 Price에 0~1의 값, 중간 가격에 2~3의 값, 높은 가격에 2~6d의 값이 분포되어 있는 것을 확인할 수 있다.
# 즉, bedroom과 bathroom의 변수를 혼합하여 새로운 변수를 도출할 수 있을 것이다.
```
